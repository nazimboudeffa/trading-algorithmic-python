# BOLLINGER BANDS
https://www.youtube.com/watch?v=5_cJs9YOPw0

```python
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = yf.download('EURUSD=X', interval='30m', period="1d")
```

```python
def bollinger_bands(df):
    rolling_mean = df['Close'].rolling(window=30).mean()
    rolling_std = df['Close'].rolling(window=30).std()
    df['UpperBand'] = rolling_mean + (2 * rolling_std)
    df['LowerBand'] = rolling_mean - (2 * rolling_std)
    return df
```

```python
df = bollinger_bands(df)
```

```python
plt.figure(figsize=(12.2,4.5))
plt.plot(df.UpperBand)
plt.plot(df.LowerBand)
plt.xticks(rotation=45)
plt.show()
```
