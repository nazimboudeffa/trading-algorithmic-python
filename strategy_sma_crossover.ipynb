```python
import yfinance as yf
import pandas as pd
from datetime import timedelta, datetime
```

```python
delta = timedelta(days=-300)
today = datetime.now()

data = yf.download("GC=F", today+delta)
```

```python
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='GOLD Close Price', alpha=0.5)
plt.plot(data['SMA_50'], label='50-Day SMA', color='blue', alpha=0.75)
plt.plot(data['SMA_200'], label='200-Day SMA', color='orange', alpha=0.75)
plt.title('GOLD Price and Moving Averages')
plt.legend()
plt.show()
```
